```python
import pandas as pd
```

```python
df = pd.read_csv('./audit_risk.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
# 从结果来看，数据没有缺失
# 但是，我们可以看到，数据中LOCATION_ID是字符串数据，我们需要将其转换为数值类型
# 我们可以使用one-hot编码的方式来处理这些数据,将其转换为数值类型
df = pd.get_dummies(df, columns=['LOCATION_ID'])
```

```python
df.head()
```

```python
# 导入决策树模型
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split

# 选取特征值和目标值
x = df.drop(['Risk'], axis=1)
y = df.loc[:, 'Risk']

# 划分训练集和测试集
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=22, test_size=0.3)

# 实例化决策树模型
dtc = DecisionTreeClassifier(random_state=22)
# 训练模型
dtc.fit(x_train, y_train)
# 预测结果
y_predict = dtc.predict(x_test)

# 模型评估
(y_predict == y_test).mean()
```

```python
# 导入决策树可视化工具
from sklearn.tree import export_graphviz
import graphviz

# 显示决策树
dot_data = export_graphviz(dtc, out_file=None, feature_names=x.columns, filled=True, rounded=True)
graph = graphviz.Source(dot_data)
graph
```
